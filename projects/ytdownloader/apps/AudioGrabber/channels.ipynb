# get all videos from the channel

```python
items = api.getAll(id='',request_item=Item.SEARCH,parts=[Part.ID,Part.SNIPPET, Part.SNIPPET.],params={'channelId':channelId})
len(items)
```

```python
items
```

```python
videoIds = []
for item in items:
    videoId = item['snippet']['resourceId']['videoId']
    data = api.getVideoData(videoId)
    if not len(data)>0:
        continue
    res = data[0]
    duration = res['duration'].total_seconds()
    if duration > 180 and duration < 660:
        videoIds.append(videoId)
len(videoIds)
```
