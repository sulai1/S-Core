```python
content,_ = api.get(id=playlistID,request_item=Item.PLAYLIST,parts=[Part.SNIPPET])
content
```

```python
title = content[0]['snippet']['title']
title
```

```python
playlistItems = api.getAll(playlistID,request_item=Item.PLAYLIST_ITEMS,parts=[Part.SNIPPET],maxResults=500)
```

```python
videoIds = []
for item in playlistItems:
    videoId = item['snippet']['resourceId']['videoId']
    data = api.getVideoData(videoId)
    if not len(data)>0:
        continue
    res = data[0]
    duration = res['duration'].total_seconds()
    if duration > 180 and duration < 660:
        videoIds.append(videoId)
len(videoIds)
```

```python
playlistFolder = os.path.join(download_folder,title)
os.mkdir(playlistFolder)
api.download(videoIds,playlistFolder)
```
