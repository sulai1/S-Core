```python
playlistItems = api.getAll(uploads,request_item=Item.PLAYLIST_ITEMS,parts=[Part.SNIPPET],maxResults=500)
```

```python
videoId = playlistItems[0]['snippet']['resourceId']['videoId']

res = api.getVideoData(videoId)

res
```

```python
videoIds = []
for item in playlistItems:
    videoId = item['snippet']['resourceId']['videoId']
    data = api.getVideoData(videoId)
    if not len(data)>0:
        continue
    res = data[0]
    duration = res['duration'].total_seconds()
    if duration > 180 and duration < 660:
        videoIds.append(videoId)
len(videoIds)
```

```python
videoIds
```

```python
videoIds = api.remove_existing(videoIds=videoIds,downloadFolder=download_folder)
len(videoIds)
```

```python
api.download(videoIds,download_folder)
```
